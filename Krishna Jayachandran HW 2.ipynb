(a) 
The polynomial regression would have a lower training RSS compared to linear regression since polynomial regression will have a tighter fit to the data contributed by the noise. (irreducible error)


(b) 
The polynomial regression would have a higher test RSS compared to linear regression because of overfitting from train data.

(c) 
The polynomial regression will have a lower train RSS than the linear regression because polynomial regression model is more flexible. Irrespective of the relationship between X and Y, the more flexible model will fit the points closer and reduce the train RSS. 

(d) 
There is not enough information to tell which test RSS would be lower because the question says "not knowing how far it is from linear." If the relationship tends more towards linear, then the test RSS for linear would be lower than the cubic regression test RSS. If the relationship is closer to cubic than linear, then the test RSS for cubic would be lower than that for linear regression. Here , the concept of bias-variance trade off kicks in and hence it can't be said what level of flexibility might fit  the data better.
